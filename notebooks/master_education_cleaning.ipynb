Plotting

```python


plt.figure(figsize=[12, 6])
for index, row in pivoted_educational_data.iterrows():
    plt.plot(row[1:].values, label=row['Location.1'])

plt.xlabel('Year')
plt.ylabel('Percent Bachelor\'s Degree or Higher')
plt.title('Time Analysis of Bachelor\'s Degree Attainment by County')
plt.legend(loc='best', fontsize='small')
plt.show()
```

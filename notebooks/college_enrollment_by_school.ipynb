```python


def plot_stats(df, year, value_column, label):
    plt.bar(df['duration'], df[value_column], label=label)
    plt.title(f'{year} Enrollment Stats by Duration')
    plt.xlabel('Duration')
    plt.ylabel('Enrollment')
    plt.legend()
    plt.show()

plot_stats(public_2016_stats, '2016', 'mean_enrollment', 'Mean Enrollment')
plot_stats(public_2016_stats, '2016', 'median_enrollment', 'Median Enrollment')
plot_stats(public_2016_stats, '2016', 'total_enrollment', 'Total Enrollment')
```

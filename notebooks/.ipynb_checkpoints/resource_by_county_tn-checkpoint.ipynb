```python
import pandas as pd
import numpy as np
import seaborn as sns
import json
import matplotlib.pyplot as plt
pd.set_option('display.max_columns', None)
```

```python
county_resources = pd.read_csv('../data/resource_mapping_by_county_2015_2022.csv')
county_resources
```

```python
county_resources.columns = [
    'year', 'agency_id', 'agency_name', 'program_name', 'county_district',
    'calculation_method', 'children_served', 'expenditure'
]
county_resources
```

```python
county_resources.dtypes
```

```python
county_resources['expenditure'] = county_resources['expenditure'].str.replace(',', '').str.replace('$', '')
county_resources['expenditure'] = pd.to_numeric(county_resources['expenditure'], errors='coerce')


print(county_resources.dtypes)

```

```python
county_resources.tail()
```

```python
print(county_resources.head())
```

```python
county_resources_DOE = county_resources[(county_resources['agency_name'] == "Department of Education")]


print(county_resources_DOE)
```

```python
county_resources_DOE['year'] = county_resources_DOE['year'].astype(int)
```

```python
DOE_2017_to_2019 = county_resources_DOE[(county_resources_DOE['year'] >= 2017) & (county_resources_DOE['year'] <= 2019)]
DOE_2017_to_2019
```

```python
DOE_2017_to_2019_estimate = DOE_2017_to_2019[DOE_2017_to_2019['calculation_method'] == 'Estimate']
DOE_2017_to_2019_estimate
```

```python
doe_2017_2019_pivot_table = DOE_2017_to_2019_estimate.pivot(index=['year', 'program_name', 'county_district'], columns='calculation_method', values='expenditure')
doe_2017_2019_pivot_table
```

```python
doe_2017_2019_pivot_table.rename(columns={'Estimate': 'estimate_expenditure'}, inplace=True)
doe_2017_2019_pivot_table
```

```python
doe_2017_2019_pivot_table.reset_index(inplace=True)
```

```python
doe_2017_2019_pivot_table
```

```python
doe_2020_to_2022 = county_resources_DOE[(county_resources_DOE['year'] >= 2020) & (county_resources_DOE['year'] <= 2022)]
doe_2020_to_2022
```

```python
doe_2020_to_2022.dtypes
```

```python
doe_2020_2022_estimates = doe_2020_to_2022[doe_2020_to_2022['calculation_method'] == 'Estimate']
doe_2020_2022_estimates
```

```python
doe_2020_2022_aggregated = doe_2020_2022_estimates.groupby(['year', 'program_name', 'county_district', 'calculation_method'])['expenditure'].sum().reset_index()
doe_2020_2022_aggregated
```

```python
doe_2020_to_2022_pivot_table = doe_2020_2022_aggregated.pivot(index=['year', 'program_name', 'county_district'], columns='calculation_method', values='expenditure')
doe_2020_to_2022_pivot_table
```

```python
doe_2020_to_2022_pivot_table_backup = doe_2020_2022_aggregated.pivot(index=['year', 'program_name', 'county_district'], columns='calculation_method', values='expenditure')
doe_2020_to_2022_pivot_table_backup
```

```python
doe_2020_to_2022_pivot_table.reset_index(inplace=True)
```

```python
collegegoing_rates_with_income = county_profiles[['county','bachelors_or_greater', 'median_income', 'collegegoingrate_2016', 'collegegoingrate_2021']]
collegegoing_rates_with_income
```

```python
collegegoing_rates_with_income = collegegoing_rates_with_income.head(95)
```

```python
collegegoing_rates_with_income
```

```python
selected_columns = [
    'collegegoingrate_2016', 'collegegoingrate_2021','bachelors_or_greater',
    'median_income'
   
]

correlation_matrix = collegegoing_rates_with_income[selected_columns].corr()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python

np.fill_diagonal(correlation_matrix.values, np.nan)

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
